```python
# prompt: i want to change the y into numeric values hateful to 1 and non hateful to 0

new_y = []
for label in y:
  if label == 'hateful':
    new_y.append(1)
  else:
    new_y.append(0)

y = new_y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0 )
```

```python
from transformers import BertTokenizer
model_name = "IMSyPP/hate_speech_en"
tokenizer = BertTokenizer.from_pretrained(model_name)
```

```python
train_encoding = tokenizer(X_train, truncation=True, padding = True)
test_encoding = tokenizer(X_test, truncation = True, padding = True)
```

```python
train_encoding
```

```python
import tensorflow as tf

train_dataset = tf.data.Dataset.from_tensor_slices((
    dict(train_encoding),
    y_train
)).batch(8)

test_dataset = tf.data.Dataset.from_tensor_slices((
    dict(test_encoding),
    y_test
)).batch(16)
```

```python
train_dataset
```

```python
test_dataset
```

```python
# prompt: i have made both the dataset test and train and now want to train by this .. i am using a hugging face model model_name = "IMSyPP/hate_speech_en"


from transformers import TFAutoModelForSequenceClassification, create_optimizer
import transformers

model = TFAutoModelForSequenceClassification.from_pretrained(model_name, from_pt = True)

optimizer, schedule = create_optimizer(
    init_lr=5e-5, num_warmup_steps=0, num_train_steps=len(train_dataset) * 3,
)

model.compile(optimizer=optimizer)

model.fit(train_dataset, epochs=3)

model.evaluate(test_dataset)
```
